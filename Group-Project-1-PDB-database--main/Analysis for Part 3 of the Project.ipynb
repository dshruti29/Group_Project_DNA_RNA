```python
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
```

```python
# Storing 3 protein  filepaths into a variable
PF_1 = Path("Resources/CSV with binding affinity of Proteins/Protein file 1.csv")
PF_2 = Path("Resources/CSV with binding affinity of Proteins/Protein file 2.csv")
PF_3 = Path("Resources/CSV with binding affinity of Proteins/Protein file 3.csv")
```

```python
PF_1_df = pd.read_csv(PF_1, header =1,usecols=["PubMed ID","Ligand","Value","Unit","Molecular Weight per Deposited Model","Stucture Keywords"])
PF_1_df
```

```python
PF_2_df = pd.read_csv(PF_2, header =1,usecols=["PubMed ID","Ligand","Value","Unit","Molecular Weight per Deposited Model","Stucture Keywords"])
PF_2_df
```

```python
PF_3_df = pd.read_csv(PF_3, header =1,usecols=["PubMed ID","Ligand","Value","Unit","Molecular Weight per Deposited Model","Stucture Keywords"])
PF_3_df
```

```python
#Combine Dataframe for anaylysis
Protein_df = pd.concat([PF_1_df, PF_2_df, PF_3_df])
Protein_df
```

```python
#Drop cells with no values
Cleaned_df = Protein_df.dropna()
Cleaned_df
```

```python
Protein_df.to_csv("Resources/CSV with binding affinity of Proteins/Protein_Combined_df.csv")
```

```python
#Identify Outliers
Binding_affinity = Cleaned_df['Value']
fig1, ax1 = plt.subplots()
ax1.set_title('Binding Affinity Box Plot')
ax1.set_ylabel('Binding Affinity(nM)')
ax1.boxplot(Binding_affinity)
plt.show()
```

```python
#Identify and show outliers

Q1 = Cleaned_df['Value'].quantile(0.25)
Q3 = Cleaned_df['Value'].quantile(0.75)
IQR = Q3 - Q1

# identify outliers
threshold = 1.5
outliers = Cleaned_df[(Cleaned_df['Value'] < Q1 - threshold * IQR) | (Cleaned_df['Value'] > Q3 + threshold * IQR)]
print(outliers)
```

```python
#Drop Outliers
New_df = Cleaned_df.drop(outliers.index)
New_df
```

```python
#Linear Regression to look for trends in size vs Kd

x_values = New_df['Molecular Weight per Deposited Model']
y_values = New_df['Value']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(10,20),fontsize=15,color="Red")
plt.xlabel('Molecular Weight(kDa)')
plt.ylabel('Binding affinity(nM)')
print(f"The r-squared is: {rvalue**2}")
plt.savefig('Images/Binding Affinity trend by size for proteins.jpg', format = 'jpg', dpi = 1200)

plt.show()
```

```python
Ligand = Path("Resources/Ligand.csv")
Ligand_df = pd.read_csv(Ligand)
Ligand_df['logKd'] = np.log10(Ligand_df['Kd'])
Ligand_df
```

```python
#Stack ligands with biomolecules
New_df = Ligand_df[['Ligand','Type','logKd']]
New_df.set_index(['Ligand','Type'])
```

```python
#Find protein ligands from Df
Protein_df_1 = New_df.loc[New_df["Type"]=="Protein"]
Protein_df_1
```

```python
#Find DNA ligands from Df

DNA_df_1 = New_df.loc[New_df["Type"]=="DNA"]
DNA_df_1
```

```python
#Find RNA ligands from Df

RNA_df_1 = New_df.loc[New_df["Type"]=="RNA"]
RNA_df_1
```

```python
#Outer Merge to show ligands sepeprated by biomolecules(Protein and DNA)
Merged_df_1 = pd.merge(Protein_df_1,DNA_df_1, on = "Ligand", how ="outer")
Merged_df_1
```

```python
#Outer Merge to show ligands sepeprated by biomolecules(All biomolecules)

Merged_df_2 = pd.merge(Merged_df_1,RNA_df_1, on = "Ligand", how ="outer")
Merged_df_2
```

```python
#Clean up biomolecule df
Merged_df_2 = Merged_df_2.drop(columns=["Type_x","Type_y","Type"])
Merged_df_2 = Merged_df_2.set_index('Ligand')
Merged_df_2=Merged_df_2.rename(columns ={'logKd_x':'logKd Protein','logKd_y':'logKd DNA','logKd':'logKd RNA'})
Merged_df_2
```

```python
#Display the ligand vs biomolecule bar chart
colors = ['green', 'purple', 'blue']
Binding_chart = Merged_df_2.plot(kind="bar", title="Binding Affinity by Biomolecule", color = colors)
plt.savefig('Images/Binding Affinity by Biomolecule.jpg', format = 'jpg', dpi = 1200)
plt.show()
```
