```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read in the csv file
project_dna_rna1 = pd.read_csv("Resources/rcsb_pdb_custom_report_66d499e3a2600d7479f8684846c186a0_0001-2500.csv", header=1)
project_dna_rna1.head()
```

```python
# Drop unnecessary columns and index
project_dna_rna1 = project_dna_rna1.drop(columns=['Structure Determination Methodology', 'Number of Distinct RNA Entities', 'Stoichiometry', 'Ligand Formula', 'Ligand MW', 'Ligand ID','Ligand Name','Unnamed: 21'])
project_dna_rna1 = project_dna_rna1.set_index('Entry ID')
project_dna_rna1.head()
```

```python
# Read in second csv
project_dna_rna2 = pd.read_csv("Resources/rcsb_pdb_custom_report_66d499e3a2600d7479f8684846c186a0_2501-4323.csv", header=1)
project_dna_rna2.head()
```

```python
# Drop unnecessary columns and index
project_dna_rna2 = project_dna_rna2.drop(columns=['Structure Determination Methodology', 'Number of Distinct RNA Entities', 'Stoichiometry', 'Ligand Formula', 'Ligand MW', 'Ligand ID','Ligand Name','Unnamed: 21'])
project_dna_rna2 = project_dna_rna2.set_index('Entry ID')
project_dna_rna2.head()
```

```python
# Concatenate files
project_dna_rna_final = pd.concat([project_dna_rna1, project_dna_rna2])
project_dna_rna_final.head()
```

```python
project_dna_rna_final.to_csv("Resources/CSV files for DNA and RNA\dna_rna_combined.csv")
```
